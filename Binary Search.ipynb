```python
## Normal Binary Search

def binary_search(lst,num):
 left_index=0
 right_index=len(lst)-1
 mid_index=0
 while left_index<=right_index:
    mid_index=(left_index+right_index)//2
    mid_number=lst[mid_index]

    if mid_number== num:
      return mid_index
    if mid_number < num:
      left_index=mid_index+1
    else:
      right_index=mid_index-1
    return -1

lst= [12,15,17,19,21,24,45,67]
print(binary_search(lst,6))
```

```python
## Recursive Binary Search

def binary_search_recursive(lst,num,left_index,right_index):
    if right_index<left_index:
        return -1
    mid_index=(left_index+right_index)//2
    mid_number=lst[mid_index]
    if mid_number==num:
        return mid_index
    if mid_number<num:
        left_index=mid_index+1
    else:
        right_index=mid_index-1
    return binary_search_recursive(lst,num,left_index,right_index)

    
lst= [12,15,17,19,21,21,21,24,45,67]
print(binary_search_recursive(lst,21,0,len(lst)-1))
```

```python
## Lower Bound Implementation
## lower bound is the largest element smaller than or equal to a given number
def findFloor(A,N,X):
        low,high,x,ans=0,N-1,X,-1
        while low<=high:
          mid=(low+high)//2
          if A[mid]==x:
              return mid
          if A[mid]>x:
            high=mid-1
          else:
            ans=mid
            low=mid+1
        return ans

print(findFloor(A=[1,2,8,10,11,12,19],N=7,X=5))
```

```python
# Find square root of number if its a perfect sqaure else return floor(sqrt(x))
## floor(sqrt(x)) = x//2
class Solution:
    def floorSqrt(self, x): 
        if x<2:
            return x
        low=2 
        high=x//2
        while low<=high:
            mid=(low+high)//2
            if mid**2==x:
                return mid
            if mid**2>x:
                high=mid-1
            else:
                low=mid+1
        return high
    
sol=Solution()
print(sol.floorSqrt(8))
```

```python
## First and Last Occurence of an Element

def first_occurence(arr,n,x):
    low,high=0,n-1
    mid=0
    while low<=high:
        mid=(low+high)//2
        if arr[mid]>=x:
            high=mid-1
        else:
            low=mid+1
    return low

def last_occurence(arr,n,x):
    low,high=0,n-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]<=x:
            low=mid+1
        else:
            high=mid-1
    return high

print(first_occurence([1,2,3,5,5,5,7,8],8,5))
print(last_occurence([1,2,3,5,5,5,7,8],8,5))
```

```python
# Rotated Sorted Array I
# Print index of element in a sorted but rotated array

def search(arr, n, k):
    low = 0
    high = n - 1
    while low <= high:
        mid = (low + high) // 2

        # if mid points the target
        if arr[mid] == k:
            return mid

        # if left part is sorted
        if arr[low] <= arr[mid]:
            if arr[low] <= k and k <= arr[mid]:
                # element exists
                high = mid - 1
            else:
                # element does not exist
                low = mid + 1
        else:  # if right part is sorted
            if arr[mid] <= k and k <= arr[high]:
                # element exists
                low = mid + 1
            else:
                # element does not exist
                high = mid - 1
    return -1
```

```python
# Search in Rotated Sorted Array II
# Given an integer array arr of size N, sorted in ascending order (may contain duplicate values) and a target value k. 
# Now the array is rotated at some pivot point unknown to you. Return True if k is present and otherwise, return False. 

def present(arr,k):
    low,high=0,len(arr)-1
    while low<=high:
        mid=(low+high)//2
        # return true if key is found
        if arr[mid]==k:
            return True
        if arr[low]==arr[mid]==arr[high]:
            low+=1
            high-=1
        elif arr[low]<=arr[mid]:
            # left part is sorted
            if arr[low]<=k and k<arr[mid]:
                high=mid-1
            else:
                low=mid+1
        else:
            # right part is sorted
            if arr[mid]<=k and k<=arr[high]:
                low=mid+1
            else:
                high=mid-1
    return False
        
arr = [5,5,5,5,4,3,3,3]
k = 4
print(present(arr,k))
```

```python
## Binary Search on Answer
## Find the Peak Element :- an element which is greater than both o its neighbours , for edge cases it should be greater than its only neighbour

def findPeakElement(arr,n):

    # Edge cases:
    if n == 1:
        return 0
    if arr[0] > arr[1]:
        return 0
    if arr[n - 1] > arr[n - 2]:
        return n - 1

    low = 1
    high = n - 2
    while low <= high:
        mid = (low + high) // 2

        # If arr[mid] is the peak:
        if arr[mid - 1] < arr[mid] and arr[mid] > arr[mid + 1]:
            return mid

        # shift to right half:
        if arr[mid] > arr[mid - 1]:
            low = mid + 1

        #shift to left
        else:
            high = mid - 1

    # Dummy return statement
    return -1


arr = [1, 2, 3, 4, 5, 6, 7, 8, 5, 1]
ans = findPeakElement(arr,len(arr))
print("The peak is at index:", ans)

```

```python
 # Find the only unique element in a sorted array all others elements repeat twice
def unique(arr):
    low,high=0,len(arr)-1
    while low<=high:
        mid=(low+high)//2
        # if we get the 2nd repetition of an element
        if arr[mid]==arr[mid-1]:
            # and we have even no. of elements on the left side
            if (mid-low+1)%2==0:
                low=mid+1
            else:
                high=mid-1
        # if we get the 1st repetition of an element
        elif arr[mid]==arr[mid+1]:
            # if we have odd no. of elements on the left side
            if (mid-low+1)%2==1:
                low=mid+1
            else:
                high=mid-1
        else:
        # return index when unique element is found
            return mid
    # return -1 if no unique element
    return -1

arr = [1,1,2,2,3,3,4,5,5,6,6]
print(unique(arr))
```

```python
# Koko Eating Bananas 
import math
def minEatingSpeed(piles: List[int], h: int) -> int:
    def check(speed,arr,h):
        sum=0
        for pile in arr:
            sum+=math.ceil(pile/speed)
        if sum<=h:
            return True
        return False
    def kokoEats(arr,h):
        low,high=1,max(arr)
        while low<=high:
            mid=(low+high)//2
            if check(mid,arr,h):
                high=mid-1
            else:
                low=mid+1
        return low
    return kokoEats(piles,h)
```

```python
# Min Days for Bouquets
def minDays( bloomDay: List[int], m: int, k: int) -> int:
    def isPossible(bloomDay,m,k,day):
        total,flowers=0,0
        for b in bloomDay:
            if b<=day:
                flowers+=1
                if flowers==k:
                    total+=1
                    flowers=0
            else:
                flowers=0
        if total>=m:
            return True
        return False

    def minDays(bloomDay,m,k):
        if m*k>len(bloomDay):
            return -1
        low=1
        high=10**9
        while low<=high:
            mid=(low+high)//2
            if isPossible(bloomDay,m,k,mid):
                high=mid-1
            else:
                low=mid+1
        return low
    return minDays(bloomDay,m,k)
```

```python
# Min Capacity of Ship 
def shipWithinDays(self, weights: List[int], days: int) -> int:
    def canCarry(weights,days,mid):
        total,currWt=0,0
        for wt in weights:
            if currWt+wt>mid:
                total+=1
                currWt=0
            currWt+=wt
        if total<=days:
            return True
        return False
    def minWeight(weights,days):
        low,high=1,sum(weights)
        while low<=high:
            mid=(low+high)//2
            if canCarry(weights,days,mid):
                high=mid-1
            else:
                low=mid+1
        return low
    return minWeight(weights,days)
```

```python
# Aggressive Cows
def canPlace(arr,minDist,cows):
    n=len(arr)
    cnt=1
    prev=arr[0]
    for i in range(1,n):
        if (arr[i]-prev)>=minDist:
            prev=arr[i]
            cnt+=1
        if cnt >= cows:
            return True
    return False
def aggresiveCows(arr,c):
    arr.sort()
    low,high=1,(arr[-1]-arr[0])
    while low<=high:
        mid=(low+high)//2
        if canPlace(arr,mid,c):
            low=mid+1
        else:
            high=mid-1
    return high
```

```python
# Book Allocation 

def canDistribute(arr,pages):
    cnt=1
    curr=0
    for i in range(len(arr)):
        if curr+arr[i]>pages:
            cnt+=1
            curr=arr[i]
        else:
            curr+=arr[i]
    return cnt

def allocate(arr,m):
    if m>len(arr):
        return -1
    maxi,sum=0,0
    for i in arr:
        sum+=i
        if i>maxi:
            maxi=i
    low,high=maxi,sum
    while low<=high:
        mid=(low+high)//2
        students=canDistribute(arr,mid)
        if students>m:
            low=mid+1
        else:
            high=mid-1
    return low
```

```python
# Painter's Partition or Split Largest Subarray Sum

def subarraySum(arr,reqSum):
    prev,cnt=0,0
    for i in range(len(arr)):
        if arr[i]+prev<=reqSum:
            prev+=arr[i]
        else:
            cnt+=1
            prev=arr[i]
    if prev>0:
        cnt+=1
    return cnt

def splitArray(arr,k):
    maxi,sum=0,0
    for i in arr:
        sum+=i
        if i>maxi:
            maxi=i
    low,high=maxi,sum
    while low<=high:
        mid=(low+high)//2
        cnt=subarraySum(arr,mid)
        if cnt>k:
            low=mid+1
        else:
            high=mid-1
    return low
```
